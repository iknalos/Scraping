```python
import time
import requests
import urllib.request
from bs4 import BeautifulSoup
from csv import DictReader
import csv
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import pandas as pd
from datetime import datetime
keywords="SQL Developer"

from rake_nltk import rake,Metric
import operator
#from nltk.corpus import stopwords
import nltk
#nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize

from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
```

```python
# To get the first 100 Jobs and extract the keywords.
i=0
total=[]
listy=[]
filtered_set=[]
for i in range(1,11,1): # iterating through first 10 pages
    try:
        dice = ('https://www.dice.com/jobs?q='+keywords+'&countryCode=US&radius=30&radiusUnit=mi&page='+str(i)+'&pageSize=20&language=en')
        options = Options()# creating an options so that the page load in background.
        options.headless = True
        browser = webdriver.Firefox(options=options)#using firefox web drivers from selenium which we imported.
        browser.get(dice)
        desc_list=browser.find_elements_by_class_name('card-title-link')
        links=[]
        for row in desc_list:# iterating through job coloumn
            x=row.get_attribute('href') # Getting links which are in 'href' tags and appending to 'links'
            links.append(x)
        for i in range(10): # iterating through first 10 jobs.
#            print (links[i])
            value=browser.get(links[i])
            view_page = browser.page_source
            job = browser.find_element_by_xpath('//*[@id="jt"]').text # extracting job title
#            print (job)
            job_description = view_page.split("<div class=\"highlight-black\" id=\"jobdescSec\">")[1].split("<div class=\"row fixedbottom\">")[0]
            job_description = str(job_description)

            lines = job_description.split('<li>') # below all lines are used to extract bullet point texts.
            y = [ x for x in lines if "</li><br>" in x ]
            y = [ x for x in lines if "<p>" not in x ]
            dice_list4 = [x.replace('</li><br>','') for x in y]
            dice_list4 = [x.replace('</b><br><br>','') for x in dice_list4]
            dice_list4 = [x.replace('<br><br><b>','') for x in dice_list4]
            dice_list4 = [x.replace('</li>','') for x in dice_list4]
            dice_list4 = [x.replace('</ul>','') for x in dice_list4]
            dice_list4 = [x.replace('<ul>','') for x in dice_list4]
            dice_list4 = [x.replace('<b>','') for x in dice_list4]
            dice_list4 = [x.replace('</b>','') for x in dice_list4]
            dice_list4 = [x.replace('<br>','') for x in dice_list4]
            dice_list4 = [x.replace('</div>','') for x in dice_list4]
            listy.append(dice_list4)
            dice_list4 = str(dice_list4)       
            tokenized_word=word_tokenize(dice_list4)# breaking sentence into words by tokenize. This is imported from nltk
            for w in tokenized_word:
                if w not in stop_words:# Comparing words with stopwords and removing which match.
                    w=w.lower()
                    filtered_set.append(w)
            fit_words= [word for word in filtered_set if word. isalnum()]

            x=nltk.pos_tag(fit_words)
            for wordz in x:
                if 'NN' in wordz[1]:
                    print (wordz)
                    total.append([job,wordz[0],wordz[1]])
           
    except IndexError: # If takes too time it ignores and continue.
        continue
```

```python
df=pd.DataFrame(total) # Inserting our final filtered_sent into panda dataframe.
    #df.to_csv("indeed_Whole.csv",header=False,index=False)  # Adding the above dataframe into nlp.csv.
df.to_csv("Dice_Noun_n.csv",header=False,index=False)  # Adding the above dataframe into nlp.csv.
```
